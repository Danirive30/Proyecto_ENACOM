# Análisis Exploratorio de Datos (ENACOM)


## Introducción

Este análisis exploratorio de datos (EDA) se centra en los servicios de internet en Argentina, utilizando datos obtenidos del Ente Nacional de Comunicaciones (ENACOM).

El objetivo de este EDA es:

1. Analizar las tendencias en el acceso a internet y su velocidad a través de diversas tecnologías.
2. Examinar la penetración de internet en la población y en los hogares.
3. Evaluar los ingresos generados por los servicios de internet.

Para lograr esto, se seleccionaron diversas hojas del dataset 'Internet.xlsx', que incluyen datos sobre accesos, velocidad, y penetración de internet.


## Hojas a utilizar

Se seleccionaron las siguientes hojas del dataset 'Internet.xlsx' para el análisis:

1. `Accesos por velocidad`: Permite analizar el número de accesos a internet por diferentes velocidades de conexión.

2. `Totales Accesos por velocidad`: Proporciona una visión general de los totales de accesos a internet desglosados por velocidad.

3. `Accesos Por Tecnología`: Ayuda a entender cómo se distribuyen las diferentes tecnologías de acceso en las provincias.

4. `Totales Accesos por Tecnología`: Este resumen facilita la comparación entre diferentes tecnologías de acceso a internet.

5. `Velocidad % por prov`: Ofrece información sobre la distribución porcentual de la velocidad de conexión a internet por provincia.

6. `Penetración-población`: Permite evaluar el porcentaje de la población con acceso a internet.

7. `Penetracion-hogares`: Similar a la penetración en la población, pero enfocado en los hogares.

8. `Ingresos`: Ofrece datos sobre los ingresos generados por los servicios de internet.

9. `Totales VMD`: La velocidad media descargada es importante para evaluar el rendimiento general del servicio de internet.

10. `Dial-BAf`: Incluye datos sobre accesos dial-up y Banda Ancha Fija, permitiendo un análisis de la evolución de las tecnologías más antiguas frente a las nuevas.

### Verificar y eliminar duplicados

```python
def verificar_duplicados(sheets):
    for name, df in sheets.items():
        duplicados_antes = df.duplicated().sum()
        df.drop_duplicates(inplace=True)
        duplicados_despues = df.duplicated().sum()
        print(f"**{name}**")
        print(f"Filas duplicadas antes: {duplicados_antes}")
        print(f"Filas duplicadas después: {duplicados_despues}\n")


# Llama a la función para verificar y eliminar duplicados en cada hoja 'Internet.xlsx'
verificar_duplicados(sheets)
```

### 5. Velocidad % por prov
Esta hoja presenta la distribución porcentual de accesos a internet por rangos de velocidad en cada provincia.


```python
sheets['Velocidad % por prov'].head()
```

```python
sheets['Velocidad % por prov']['Provincia'].unique()
```

```python
# Verificar estadísticas descriptivas para la columna de velocidad
print(sheets['Velocidad % por prov']['Mbps (Media de bajada)'].describe())
```

```python
# Calcular el rango intercuartílico (IQR)
Q1 = sheets['Velocidad % por prov']['Mbps (Media de bajada)'].quantile(0.25)
Q3 = sheets['Velocidad % por prov']['Mbps (Media de bajada)'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = sheets['Velocidad % por prov'][(sheets['Velocidad % por prov']['Mbps (Media de bajada)'] < lower_bound) |
                                          (sheets['Velocidad % por prov']['Mbps (Media de bajada)'] > upper_bound)]

print(f'Número de outliers: {outliers.shape[0]}')
print(outliers)
```

```python
# Visualizar outliers utilizando un gráfico de caja
plt.figure(figsize=(10, 6))
sns.boxplot(data=sheets['Velocidad % por prov'],
            y='Mbps (Media de bajada)', color='lightblue')
plt.title('Boxplot de la Velocidad Media de Bajada por Provincia')
plt.ylabel('Mbps (Media de bajada)')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Gráfico de Barras de Mbps (Media de bajada) por Provincia
plt.figure(figsize=(12, 6))
sns.barplot(data=sheets['Velocidad % por prov'], x='Provincia',
            y='Mbps (Media de bajada)', palette='viridis')
plt.title('Mbps (Media de bajada) por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Mbps (Media de bajada)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

De lo anterior se puede concluir que las provincias con menor velocidad de bajada son Chubut, La Pampa, San Juan, Santa Cruz y Tierra del Fuego. Por otro lado, las provincias con mayor velocidad de bajada son Capital Federal, Buenos Aires y Córdoba.

### data_general

```python
# Se realiza un merge de los dataframes 'data_general' limpiados para tenerlos en un solo archivo

data_general = pd.merge(sheets['Totales Accesos Por Tecnología'],
                        sheets['Totales Accesos por velocidad'], on=['Año', 'Trimestre'], how='inner')
data_general = pd.merge(data_general, sheets['Ingresos'], on=[
                        'Año', 'Trimestre'], how='inner')
data_general = pd.merge(data_general, sheets['Totales VMD'], on=[
                        'Año', 'Trimestre'], how='inner')
data_general.head()
```

### data_provincias

```python
# Se realiza un merge de los dataframes 'data_provincias' limpiados para tenerlos en un solo archivo

data_provincias = pd.merge(sheets['Accesos por velocidad'], sheets['Accesos Por Tecnologia'], on=[
                           'Año', 'Trimestre', 'Provincia'], how='inner')
data_provincias = pd.merge(data_provincias, sheets['Velocidad % por prov'], on=[
                           'Año', 'Trimestre', 'Provincia'], how='inner')
data_provincias = pd.merge(data_provincias, sheets['Penetración-poblacion'], on=[
                           'Año', 'Trimestre', 'Provincia'], how='inner')
data_provincias = pd.merge(data_provincias, sheets['Penetracion-hogares'], on=[
                           'Año', 'Trimestre', 'Provincia'], how='inner')
data_provincias = pd.merge(
    data_provincias, sheets['Dial-BAf'], on=['Año', 'Trimestre', 'Provincia'], how='inner')
data_provincias.head()
```
